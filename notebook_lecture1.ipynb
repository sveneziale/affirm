### Example 1: Division algorithm 

In one variable, the division algorithm is unique: given $f, g \in k[x]$, there exist unique $q, r$ with $f = qg + r$ and $\deg(r) < \deg(g)$.

```python
# This is how we define a polynomial ring in SageMath (with a single variable x over the rational numbers QQ -- you can also over RR, but then the numbers look bad)
R.<x> = QQ[]

# Define two polynomials, f and g
f = x^5 + 2*x^3 - x + 7
g = x^2 + 1

# Run the division algorithm 
q, r = f.quo_rem(g)

print(f"Check: f == q*g + r? {f == q*g + r}")
print(f"deg(r) = {r.degree()} < deg(g) = {g.degree()}? {r.degree() < g.degree()}")
```
